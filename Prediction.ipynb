```python
import pandas as pd
import joblib
```

```python
final_model=joblib.load('Machine_efficiency.pkl')
print('Model Loaded Successfully')
```

```python
test_input=pd.DataFrame(
    [[1.45, 0.82, 92.5, 18.7, 0.18, 0.14, 0.19]],
    columns=[
        'vibration',
        'acoustic',
        'temperature',
        'current',
        'IMF_1',
        'IMF_2',
        'IMF_3'
    ]
)
```

```python
y_pred=final_model.predict(test_input)
```

```python
allocation_columns=['machine_status']
y_pred_name=pd.DataFrame(y_pred,columns=allocation_columns)
label='Normal Condition' if y_pred_name.iloc[0,0] == 0 else 'Fault / Degradation'
print(f'Status Of Machine is={label}')
```
